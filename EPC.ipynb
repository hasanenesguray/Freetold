```python
import requests
from bs4 import BeautifulSoup
import time
import random
import pandas as pd
```

```python
postcodes_df = pd.read_csv('postcodes.csv')
postcodes_df
```

```python
postcodes = postcodes_df['postcode'].tolist()
postcodes
```

```python
urls = list()
```

```python
for postcode in range(len(postcodes)):
    # URL of the page you want to scrape
    url = 'https://find-energy-certificate.service.gov.uk/find-a-certificate/search-by-postcode?postcode=' + postcodes[postcode]

    # Headers to mimic a browser visit
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
    }

    # Send a GET request to the URL with headers
    response = requests.get(url, headers=headers)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all links with the class 'govuk-link' and 'rel' attribute 'nofollow'
        links = soup.find_all('a', class_='govuk-link', rel='nofollow')

        # List to store the hrefs
        hrefs = []

        # Loop through each link and extract the href attribute
        for link in links:
            hrefs.append('https://find-energy-certificate.service.gov.uk/' + link.get('href'))

    else:
        print("Failed to retrieve the webpage. Status code:", response.status_code)
    
    urls += hrefs
    print(str(postcode+1)," / ",str(len(postcodes)))
    time.sleep(random.uniform(1, 5))
    
```

```python
epc_list = list()
```

```python
for url in range(594,len(urls)):
    response = requests.get(urls[url], headers=headers)

    soup = BeautifulSoup(response.text, 'html.parser')

    # energy_rating
    energy_rating_element = soup.find("p", class_="epc-rating-result")
    energy_rating = energy_rating_element.text.strip() if energy_rating_element else "Rating not found"

    # valid_until
    valid_until_element = soup.find("div", class_="epc-extra-box").find("p", class_="govuk-body govuk-!-font-weight-bold")
    valid_until = valid_until_element.text.strip() if valid_until_element else "Valid date not found"

    # certificate_number
    certificate_number_element = soup.find_all("p", class_="govuk-body govuk-!-font-weight-bold")[1]  # Assuming it's the second occurrence
    certificate_number = certificate_number_element.text.strip() if certificate_number_element else "Certificate number not found"

    # property_type
    property_type_div = soup.find('div', class_='govuk-summary-list__row')
    property_type_key = property_type_div.find('dt', class_='govuk-summary-list__key').text.strip()
    property_type = property_type_div.find('dd', class_='govuk-summary-list__value').text.strip()

    # total_floor_area
    total_floor_area = soup.find('dt', string='Total floor area').find_next_sibling('dd').text.strip()

    # current_energy_rating
    svg = soup.find("svg", class_="rating-current rating-label")
    current_energy_rating = svg.text.strip()

    # potential_energy_rating
    potential_rating_svg = soup.find('svg', class_='rating-potential rating-label')
    potential_energy_rating = potential_rating_svg.text.strip() if potential_rating_svg else "Not found"

    features_dict = {}

    tbody = soup.find('tbody', class_='govuk-table__body')

    for row in tbody.find_all('tr', class_='govuk-table__row'):
        feature = row.find('th', class_='govuk-table__cell').text.strip()
        description = row.find_all('td', class_='govuk-table__cell')[0].text.strip()
        rating = row.find_all('td', class_='govuk-table__cell')[1].text.strip()

        # Add the details to the dictionary
        features_dict[feature] = {'Description': description, 'Rating': rating}

    wall_description = features_dict.get('Wall', {}).get('Description', 'N/A')
    wall_rating = features_dict.get('Wall', {}).get('Rating', 'N/A')

    window_description = features_dict.get('Window', {}).get('Description', 'N/A')
    window_rating = features_dict.get('Window', {}).get('Rating', 'N/A')

    main_heating_description = features_dict.get('Main heating', {}).get('Description', 'N/A')
    main_heating_rating = features_dict.get('Main heating', {}).get('Rating', 'N/A')

    main_heating_control_description = features_dict.get('Main heating control', {}).get('Description', 'N/A')
    main_heating_control_rating = features_dict.get('Main heating control', {}).get('Rating', 'N/A')

    hot_water_description = features_dict.get('Hot water', {}).get('Description', 'N/A')
    hot_water_rating = features_dict.get('Hot water', {}).get('Rating', 'N/A')

    lighting_description = features_dict.get('Lighting', {}).get('Description', 'N/A')
    lighting_rating = features_dict.get('Lighting', {}).get('Rating', 'N/A')

    roof_description = features_dict.get('Roof', {}).get('Description', 'N/A')
    roof_rating = features_dict.get('Roof', {}).get('Rating', 'N/A')

    floor_description = features_dict.get('Floor', {}).get('Description', 'N/A')
    floor_rating = features_dict.get('Floor', {}).get('Rating', 'N/A')

    secondary_heating_description = features_dict.get('Secondary heating', {}).get('Description', 'N/A')
    secondary_heating_rating = features_dict.get('Secondary heating', {}).get('Rating', 'N/A')

    # primary_energy_use
    primary_energy_use_tag = soup.find('h3', text='Primary energy use')
    primary_energy_use_paragraph = primary_energy_use_tag.find_next('p')
    primary_energy_use = primary_energy_use_paragraph.get_text()

    #energy_bills
    energy_bills_div = soup.find('div', id='bills-affected')
    text_paragraph = energy_bills_div.get_text(separator=' ', strip=True)
    energy_bills = ' '.join(text_paragraph.split())

    # environmental_impact
    environmental_impact_div = soup.find('div', id='energy')
    extracted_text = ""
    if environmental_impact_div:
        # Iterate through the desired elements within the div
        for element in environmental_impact_div.find_all(['p', 'h3', 'dt', 'dd'], recursive=True):
            # Append each element's text to the string, with space for separation
            extracted_text += element.get_text(strip=True) + " "
    environmental_impact = ' '.join(extracted_text.split())

    # address
    address_p = soup.find('p', class_='epc-address')
    address = address_p.get_text(separator=' ', strip=True)


    # Saving the extracted information into a dictionary
    energy_rating_info = {
        "address": address,
        "energy_rating": energy_rating,
        "certificate_number": certificate_number,
        "valid_until": valid_until,
        "property_type": property_type,
        "total_floor_area": total_floor_area,
        "current_energy_rating": current_energy_rating,
        "potential_energy_rating": potential_energy_rating,
        "wall_description": wall_description,
        "wall_rating": wall_rating,
        "window_description": window_description,
        "window_rating": window_rating,
        "main_heating_description": main_heating_description,
        "main_heating_rating": main_heating_rating,
        "main_heating_control_description": main_heating_control_description,
        "main_heating_control_rating": main_heating_control_rating,
        "hot_water_description": hot_water_description,
        "hot_water_rating": hot_water_rating,
        "lighting_description": lighting_description,
        "lighting_rating": lighting_rating,
        "roof_description": roof_description,
        "roof_rating": roof_rating,
        "floor_description": floor_description,
        "floor_rating": floor_rating,
        "secondary_heating_description": secondary_heating_description,
        "secondary_heating_rating": secondary_heating_rating,
        "primary_energy_use": primary_energy_use,
        "energy_bills": energy_bills,
        "environmental_impact": environmental_impact
    }
    epc_list.append(energy_rating_info)
    print(str(url+1)," / ",str(len(urls)))
    time.sleep(random.uniform(1, 5))
    
epc_list
```
